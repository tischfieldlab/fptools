```python
import tdt
import pandas as pd
import glob
import os
```

```python
# Set the following variables.

# TANK_PATH: the directory path to search for blocks. The search will be recursive, so specify the top-most folder for blocks you want to inventory
TANK_PATH = r'M:\2024 July,Aug GRABDA2m data (DLS C1906Y)\Progressive Ratio tanks'

# INVENTORY_DEST: the destination for the saved inventory file. Should have a file extension of '.tsv'.
# By default it will be saved to a file named 'inventory.tsv' in the `TANK_PATH` directory, but you can change this if you wish.
INVENTORY_DEST = os.path.join(TANK_PATH, 'inventory.tsv')
```

```python
results = []
for p in glob.glob(os.path.join(TANK_PATH, '**/*.tbk'), recursive=True):
    block_dir = os.path.dirname(p)
    block = tdt.read_block(block_dir, evtype=[])
    results.append(dict(block.info.items()))
results = pd.DataFrame(results)
results.to_csv(INVENTORY_DEST, sep='\t', index=False)
results
```
