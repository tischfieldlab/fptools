```python
def to_numpy_array(arr):
    if isinstance(arr, np.ndarray):
        return arr
    elif isinstance(arr, (list, tuple)):
        return np.array(arr)
    elif len(arr)==0:
        return np.nan  # Empty array if not convertible

testdf['mean binned stream'] = testdf['mean binned stream'].apply(to_numpy_array)
```

```python
def handle_nan_arrays(arr):
    return np.array(arr) if isinstance(arr, (list, np.ndarray)) else np.nan

testdf['mean binned stream'] = testdf['mean binned stream'].apply(handle_nan_arrays)
```
